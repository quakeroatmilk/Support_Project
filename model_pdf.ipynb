```python
import spacy

# Load the model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "AL MUHRA SDN. BHD. was incorporated on 06-12-2016 in Kuching, Sarawak."

# Process the text to extract named entities
doc = nlp(text)

# Print recognized named entities
for ent in doc.ents:
    print(ent.text, ent.label_)

```

```python
import PyPDF2
import spacy

# Load spaCy's pre-trained NER model
nlp = spacy.load("en_core_web_sm")

# Function to read text from a PDF file
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            text += page.extract_text()
    return text

# Provide the path to your PDF file
pdf_path = r"C:\Users\BLESS - AZLINA\Desktop\support_entre\support_project\data\SSM.pdf"

# Extract text from the PDF
pdf_text = extract_text_from_pdf(pdf_path)

# Print the extracted text (Optional: for verification)
print("Extracted Text from PDF:\n", pdf_text)

# Perform Named Entity Recognition (NER) on the extracted text
doc = nlp(pdf_text)

# Extract and print named entities (like ORG for organizations, GPE for locations, and DATE for dates)
for ent in doc.ents:
    print(f"{ent.text}: {ent.label_}")
```
